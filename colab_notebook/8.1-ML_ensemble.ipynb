**Ensemble learning**

Son técnicas de aprendizaje automático en las cuales se combinan varios modelos para generar un modelo predictivo más óptimo.

Las cuatro clases principales métodos de Ensemble learning son: **voting, bagging, stacking y boosting**

- *Bagging* implica ajustar muchos árboles de decisión en diferentes muestras del mismo conjunto de datos y promediar las predicciones.
- *Stacking* implica ajustar muchos tipos de modelos diferentes en los mismos datos y usar otro modelo para aprender a combinar mejor las predicciones.
- *Boosting* implica agregar miembros del conjunto secuencialmente que corrigen las predicciones hechas por modelos anteriores y genera un promedio ponderado de las predicciones.
- *Voting* implica hacer una predicción que es el promedio de muchos otros modelos. Un conjunto de *votación dura* implica sumar los votos para etiquetas de clase de otros modelos y predecir la clase con la mayor cantidad de votos. Un conjunto de *votación suave* implica sumar las probabilidades predichas para las etiquetas de clase y predecir la etiqueta de clase con la mayor probabilidad de suma.

## **Stacking Ensemble Learning**

Es un procedimiento general en el que se entrena a cada modelo de forma independiente para combinar a los *individual learners*. Aquí, los *individual learners* se denominan  first-level learners, mientras que el combinador se denomina *second-level learner*, o *meta-learner*.

Podemos resumir los elementos clave:

- Conjunto de datos de entrenamiento sin cambios.
- Diferentes algoritmos de aprendizaje automático para cada miembro del ensemble.
- Modelo de aprendizaje automático para aprender a combinar mejor las predicciones.

Es deseable utilizar un conjunto de modelos que se aprendan o construyan de maneras muy diferentes, asegurándose de que hagan suposiciones diferentes y, a su vez, tengan errores de predicción menos correlacionados.

## Bagging: reducción de varianza

Si $$\operatorname{Var}[f^{(m)}(x)]=\sigma^2$$ y $$\operatorname{Cov}[f^{(i)}(x),f^{(j)}(x)]=\rho\sigma^2$$ para $i\neq j$,
$$
\operatorname{Var}[\hat{f}_{\text{bag}}(x)] = \frac{\sigma^2}{M}(1-\rho) + \rho\sigma^2.
$$

Mostramos cómo varía la varianza con $M$ para distintos $\rho$.


```r
# Gráfico Var vs M para distintos rho
sigma2 = 1.0
Ms = np.arange(1,201)
rhos = [0.0, 0.2, 0.5]
plt.figure(figsize=(7,4))
for rho in rhos:
    var = (sigma2/Ms)*(1-rho) + rho*sigma2
    plt.plot(Ms, var, label=f"rho={rho}")
plt.xlabel("M (número de modelos)")
plt.ylabel("Varianza del ensemble")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('figures/var_vs_M.png', dpi=150)
plt.show()
```

## OOB estimate vs k-fold CV (RandomForest)

### Out‑of‑bag (OOB) — definición y fórmulas

Sea el conjunto de entrenamiento
$$
\mathcal{D}=\{(x_i,y_i)\}_{i=1}^n.
$$

Para un ensemble de $M$ estimadores entrenados por bootstrap, definimos para cada observación $i$ el conjunto de estimadores que **no** la incluyeron en su muestra bootstrap:
$$
S_i=\{m\in\{1,\dots,M\}:\ (x_i,y_i)\notin\mathcal{D}^{(m)}\},
$$
donde $\mathcal{D}^{(m)}$ es la muestra bootstrap usada para entrenar el estimador $f^{(m)}$.

La predicción OOB para la observación $i$ se define como el promedio (o voto) de los estimadores en $S_i$:

$$\hat{y}_i^{\mathrm{OOB}}=
\begin{cases}
\displaystyle\frac{1}{|S_i|}\sum_{m\in S_i} f^{(m)}(x_i), & |S_i|>0,\\
\text{no disponible (usar otra estrategia)}, & |S_i|=0.
\end{cases}$$

La estimación OOB de la pérdida (error) sobre el conjunto de entrenamiento es
$$
\mathrm{Err}_{\mathrm{OOB}}=\frac{1}{n}\sum_{i=1}^n L\big(y_i,\hat{y}_i^{\mathrm{OOB}}\big),
$$
donde \(L\) es la función de pérdida (por ejemplo, \(L\) = 0–1 para clasificación o $L=(y-\hat y)^2$ para regresión).

**Propiedades e intuición breve**

- En promedio, una muestra bootstrap deja fuera aproximadamente una fracción $1/e\approx 0.37$ de las observaciones, por lo que cada observación suele ser OOB para varios estimadores.  
- OOB proporciona una estimación interna del error sin necesidad de particionar adicionalmente los datos; con suficientes estimadores suele aproximar bien el error de test.  
- Limitaciones: no es apropiado para datos con dependencia temporal y puede ser optimista si hay fugas de información en el preprocesado.

**Nota práctica (sklearn)**

En `scikit-learn` se activa con `oob_score=True` en `RandomForestClassifier`/`RandomForestRegressor`. El atributo resultante `oob_score_` devuelve la puntuación OOB (por ejemplo, accuracy para clasificación).


Definición OOB: para cada i, usar solo los estimadores que no vieron i en su bootstrap.
$$
\hat{y}_i^{\text{OOB}} = \frac{1}{|S_i|}\sum_{m\in S_i} f^{(m)}(x_i).
$$
La estimación OOB del error es
$$
\text{Err}_{\text{OOB}} = \frac{1}{n}\sum_{i=1}^n L\big(y_i,\hat{y}_i^{\text{OOB}}\big).
$$


```r
# Cargar dataset Digits y comparar OOB vs CV
from sklearn.datasets import load_digits
X, y = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RND, stratify=y)

# RandomForest con OOB
rf = RandomForestClassifier(n_estimators=200, oob_score=True, random_state=RND, n_jobs=-1)
rf.fit(X_train, y_train)
oob_score = rf.oob_score_
cv_score = np.mean(cross_val_score(RandomForestClassifier(n_estimators=200, random_state=RND), X_train, y_train, cv=5, n_jobs=-1))
test_acc = rf.score(X_test, y_test)

print(f"OOB score (train): {oob_score:.4f}")
print(f"CV 5-fold (train): {cv_score:.4f}")
print(f"Test accuracy (holdout): {test_acc:.4f}")
```

## Stacking seguro (sin fuga) - protocolo k-fold

Procedimiento:
1. Particionar en $K$ folds.
2. Para cada base learner y cada fold $t$: entrenar en $\mathcal{D}\setminus\mathcal{D}_t$ y predecir en $\mathcal{D}_t$.
3. Construir la matriz $Z\in\mathbb{R}^{n\times K_{\text{models}}}$ con predicciones nivel-0.
4. Entrenar meta-modelo $g$ sobre $(Z,y)$.
5. Para predecir en nuevos datos, reentrenar cada base learner en todo $\mathcal{D}$ y usar $g(f_1(x^\star),\dots,f_K(x^\star))$.


```r
# Implementación: función que genera predicciones nivel-0 por CV
def generate_level0_features(X, y, base_learners, n_splits=5, random_state=RND):
    n = X.shape[0]
    K = len(base_learners)
    Z = np.zeros((n, K))
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    for fold_idx, (train_idx, val_idx) in enumerate(kf.split(X)):
        X_tr, X_val = X[train_idx], X[val_idx]
        y_tr = y[train_idx]
        for j, (name, model) in enumerate(base_learners):
            from sklearn.base import clone
            m = clone(model)
            m.fit(X_tr, y_tr)
            Z[val_idx, j] = m.predict(X_val)
    return Z

# Ejemplo con 3 base learners
base_learners = [
    ("rf", RandomForestClassifier(n_estimators=100, random_state=RND)),
    ("gb", GradientBoostingClassifier(n_estimators=100, random_state=RND)),
    ("xgb", xgb.XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='mlogloss', random_state=RND))
]

Z = generate_level0_features(X_train, y_train, base_learners, n_splits=5)
print("Shape Z:", Z.shape)

# Entrenar meta-modelo (logistic regression sobre predicciones)
meta = LogisticRegression(max_iter=1000, random_state=RND)
meta.fit(Z, y_train)

# Predicción en test: reentrenar base learners en todo X_train
from sklearn.base import clone
base_trained = []
for name, model in base_learners:
    m = clone(model)
    m.fit(X_train, y_train)
    base_trained.append((name, m))

# Construir features nivel-0 para X_test
Z_test = np.column_stack([m.predict(X_test) for _, m in base_trained])
y_pred = meta.predict(Z_test)
print("Stacking test accuracy:", accuracy_score(y_test, y_pred))
```

## Boosting — explicación de términos

**Modelo aditivo**  
$$
\hat F_M(x)=\sum_{m=1}^M \nu\,\gamma_m\,h_m(x)
$$

- **$\hat F_M(x)$**: predicción final del ensemble después de $M$ iteraciones.  
- **$M$**: número total de estimadores base (iteraciones).  
- **$\nu$**: *learning rate* (tasa de aprendizaje), $0<\nu\leq 1$; reduce el aporte de cada nuevo componente.  
- **$\gamma_m$**: peso escalar del $m$-ésimo estimador; en gradient boosting se obtiene minimizando la pérdida en la dirección de $h_m$.  
- **$h_m(x)$**: $m$-ésimo *base learner* (por ejemplo, un árbol poco profundo) ajustado en la iteración $m$.

**Pseudo‑residuos (gradientes)**  
$$
r_i^{(m)} = -\left.\frac{\partial L(y_i,F(x_i))}{\partial F(x_i)}\right|_{F=\hat F_{m-1}}
$$

- **$r_i^{(m)}$**: pseudo‑residuo (gradiente negativo) para la observación $i$ en la iteración $m$; indica la dirección en la que $h_m$ debe corregir la predicción.  
- **$L(y,F(x))$**: función de pérdida (por ejemplo, log‑loss para clasificación, MSE para regresión).  
- **$\hat F_{m-1}$**: predicción agregada hasta la iteración $m-1$; la derivada se evalúa en este punto.

**Cómo se obtiene $\gamma_m$**

En general se elige $\gamma_m$ resolviendo
$$
\gamma_m \;=\; \arg\min_{\gamma}\;\sum_{i=1}^n L\big(y_i,\; \hat F_{m-1}(x_i) + \gamma\,h_m(x_i)\big)
$$

**Ejemplo concreto: pérdida cuadrática**

Si $L(y,F)=(y-F)^2$, entonces
$$
r_i^{(m)} = -\frac{\partial}{\partial F}(y_i-F)^2\Big|_{F=\hat F_{m-1}} = 2\big(y_i-\hat F_{m-1}(x_i)\big)
$$

En este caso $h_m$ se ajusta a los residuos $y_i-\hat F_{m-1}(x_i)$ y $\gamma_m$ suele tomarse como $1$ o estimarse por regresión entre $h_m(x)$ y los residuos.

**Notas prácticas**

- Reducir $\nu$ (p. ej. $0.01$–$0.1$) mejora la robustez frente a overfitting pero requiere aumentar $M$.  
- Además

```r
# Experimento: variar learning rate en GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
lrs = [0.01, 0.05, 0.1, 0.2]
results = {}
for lr in lrs:
    gb = GradientBoostingClassifier(n_estimators=200, learning_rate=lr, random_state=RND)
    scores = cross_val_score(gb, X_train, y_train, cv=5, n_jobs=-1)
    results[lr] = scores
    print(f"lr={lr}: mean acc={scores.mean():.4f}, std={scores.std():.4f}")

# Guardar figura resumen
plt.figure(figsize=(6,4))
plt.errorbar(lrs, [results[lr].mean() for lr in lrs], yerr=[results[lr].std() for lr in lrs], marker='o')
plt.xscale('log')
plt.xlabel('learning rate')
plt.ylabel('CV accuracy')
plt.grid(True)
plt.tight_layout()
plt.savefig('figures/gb_lr_vs_acc.png', dpi=150)
plt.show()
```

## Desafío A: comparar RandomForest vs XGBoost en Digits

Métricas: accuracy, tiempo de entrenamiento, estabilidad (std en repeated CV).


```r
import time
from sklearn.model_selection import RepeatedKFold

rkf = RepeatedKFold(n_splits=5, n_repeats=3, random_state=RND)
models = {
    "RandomForest": RandomForestClassifier(n_estimators=200, random_state=RND, n_jobs=-1),
    "XGBoost": xgb.XGBClassifier(n_estimators=200, use_label_encoder=False, eval_metric='mlogloss', random_state=RND, n_jobs=-1)
}

summary = []
for name, model in models.items():
    t0 = time.time()
    scores = cross_val_score(model, X_train, y_train, cv=rkf, n_jobs=-1)
    t1 = time.time()
    summary.append({
        "model": name,
        "mean_acc": scores.mean(),
        "std_acc": scores.std(),
        "time_sec": t1 - t0
    })
pd.DataFrame(summary)
```
