**Ensemble learning**

Son técnicas de aprendizaje automático en las cuales se combinan varios modelos para generar un modelo predictivo más óptimo.

Las cuatro clases principales métodos de Ensemble learning son: **voting, bagging, stacking y boosting**

- *Bagging* implica ajustar muchos árboles de decisión en diferentes muestras del mismo conjunto de datos y promediar las predicciones.
- *Stacking* implica ajustar muchos tipos de modelos diferentes en los mismos datos y usar otro modelo para aprender a combinar mejor las predicciones.
- *Boosting* implica agregar miembros del conjunto secuencialmente que corrigen las predicciones hechas por modelos anteriores y genera un promedio ponderado de las predicciones.
- *Voting* implica hacer una predicción que es el promedio de muchos otros modelos. Un conjunto de *votación dura* implica sumar los votos para etiquetas de clase de otros modelos y predecir la clase con la mayor cantidad de votos. Un conjunto de *votación suave* implica sumar las probabilidades predichas para las etiquetas de clase y predecir la etiqueta de clase con la mayor probabilidad de suma.

## **Bagging Ensemble Learning**

El nombre **Bagging** proviene de la abreviatura de **Bootstrap AGGregatING**. Como su nombre lo indica, los dos ingredientes clave de Bagging son el Bootstrap y la agregación.

Funciona de la siguiente manera:

- Se toman muestras de [Bootstrap](https://en.wikipedia.org/wiki/Bootstrapping_(statistics)) del conjunto de datos de entrenamiento.
- Los árboles de decisión no podados caben en cada muestra.
- Votación simple o promedio de predicciones.

## **Stacking Ensemble Learning**

Es un procedimiento general en el que se entrena a cada modelo de forma independiente para combinar a los *individual learners*. Aquí, los *individual learners* se denominan  first-level learners, mientras que el combinador se denomina *second-level learner*, o *meta-learner*.

Podemos resumir los elementos clave:

- Conjunto de datos de entrenamiento sin cambios.
- Diferentes algoritmos de aprendizaje automático para cada miembro del ensemble.
- Modelo de aprendizaje automático para aprender a combinar mejor las predicciones.

Es deseable utilizar un conjunto de modelos que se aprendan o construyan de maneras muy diferentes, asegurándose de que hagan suposiciones diferentes y, a su vez, tengan errores de predicción menos correlacionados.

Algunos algoritmos basados en este enfoque son:

- Stacked Models (canonical stacking)
- Blending
- Super Ensemble

## **Boosting Ensemble Learning**

En este método, el conjunto de datos de entrenamiento para cada clasificador posterior se enfoca cada vez más en instancias mal clasificadas por clasificadores generados previamente.

— Page 13, [Ensemble Machine Learning](https://www.amazon.com/Ensemble-Machine-Learning-Methods-Applications/dp/1441993258/ref=as_li_ss_tl?dchild=1&keywords=ensemble+learning&qid=1592167152&s=books&sr=1-12&linkCode=sl1&tag=inspiredalgor-20&linkId=cd7eb4abb7c18cbebb8582798232e1d3&language=en_US), 2012.

Por lo general, el conjunto de datos de entrenamiento no se modifica y, en su lugar, el algoritmo de aprendizaje se modifica para prestar más o menos atención a ejemplos específicos (filas de datos) en función de si los miembros del ensemble agregados previamente los predijeron correcta o incorrectamente. Por ejemplo, las filas de datos se pueden pesar para indicar la cantidad de enfoque que debe dar un algoritmo de aprendizaje mientras aprende el modelo.

**Ejemplo de clases sesgadas por el peso:** https://colab.research.google.com/drive/1-DKcdUI44Vv33218Aphyxpfek-HLUToj?usp=sharing

Podemos resumir los elementos clave:

- Sesgar los datos de entrenamiento hacia aquellos ejemplos que son difíciles de predecir.
- Agregue iterativamente miembros del conjunto para corregir las predicciones de modelos anteriores.
- Combine predicciones usando un promedio ponderado de modelos.

<img src='https://machinelearningmastery.com/wp-content/uploads/2020/11/Boosting-Ensemble.png'>

Fuente: https://machinelearningmastery.com/tour-of-ensemble-learning-algorithms/

Algunos algoritmos populares que se basan en ese enfoque son:

- AdaBoost (canonical boosting)
- Gradient Boosting Machines
- Stochastic Gradient Boosting (XGBoost and similar)
